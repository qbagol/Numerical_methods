```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix=np.zeros((100,100))

u0=10

for i in range(20,40):
    matrix[i,60]=u0
    matrix[i,80]=-u0

plt.imshow(matrix)            
```

```python
matrix1=np.zeros((100,100))

for x in range(100):
    for i in range(0,99):
        for j in range(0,99):
            if matrix[i,j]!=0:
                matrix1[i,j]=matrix[i,j]
            else:
                matrix1[i,j]=(matrix1[i-1,j]+matrix1[i-1,j-1]+matrix1[i-1,j+1]+matrix1[i+1,j]+matrix1[i+1,j+1]+matrix1[i+1,j-1]+matrix1[i,j+1]+matrix1[i,j-1])/8


plt.imshow(matrix1)
```
