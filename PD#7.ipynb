```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
la1,la2,la3=1, 0.1, 0.11
lb1,lb2,lb3=0.1, 1, 0.1

c1,c2,c3,c4=1/6, 1/3, 1/3 ,1/6
a2,a3,a4=1/2,1/2,1
b21,b31,b32,b41,b42,b43=1/2,0,1/2,0,0,1

timemax=60
dt=0.001
Na=1000
Nb=0
Nc=0
i=1

ttab=np.arange(0,timemax, dt)

Natab1=np.zeros(len(ttab))
Nbtab1=np.zeros(len(ttab))
Nctab1=np.zeros(len(ttab))
Natab2=np.zeros(len(ttab))
Nbtab2=np.zeros(len(ttab))
Nctab2=np.zeros(len(ttab))
Natab3=np.zeros(len(ttab))
Nbtab3=np.zeros(len(ttab))
Nctab3=np.zeros(len(ttab))
Natab1[0]=Na
Nbtab1[0]=Nb
Nctab1[0]=Nc
Natab2[0]=Na
Nbtab2[0]=Nb
Nctab2[0]=Nc
Natab3[0]=Na
Nbtab3[0]=Nb
Nctab3[0]=Nc

def dNa1(Na):
    return -la1*Na
def dNb1(Na, Nb):
    return la1*Na-lb1*Nb
def dNc1(Nb):
    return lb1*Nb
def dNa2(Na):
    return -la2*Na
def dNb2(Na, Nb):
    return la2*Na-lb2*Nb
def dNc2(Nb):
    return lb2*Nb
def dNa3(Na):
    return -la3*Na
def dNb3(Na, Nb):
    return la3*Na-lb3*Nb
def dNc3(Nb):
    return lb3*Nb

for i in range(len(ttab)):
    if Natab1[i-1]>0 or Nbtab1[i-1]>0:
        K1a=dNa1(Natab1[i-1])
        K2a=dNa1(Natab1[i-1]+a2*dt)
        K3a=dNa1(Natab1[i-1]+a3*dt)
        K4a=dNa1(Natab1[i-1]+a4*dt)
        K1b=dNb1(Natab1[i-1],       Nbtab1[i-1])
        K2b=dNb1(Natab1[i-1]+a2*dt, Nbtab1[i-1]+b21*K1b*dt)
        K3b=dNb1(Natab1[i-1]+a3*dt, Nbtab1[i-1]+b31*K1b*dt+b32*K2b*dt)
        K4b=dNb1(Natab1[i-1]+a4*dt, Nbtab1[i-1]+b41*K1b*dt+b42*K2b+b43*K3b*dt)   
        K1c=dNc1(Nbtab1[i-1])
        K2c=dNc1(Nbtab1[i-1]+a2*dt)
        K3c=dNc1(Nbtab1[i-1]+a3*dt)
        K4c=dNc1(Nbtab1[i-1]+a4*dt)

        Natab1[i]=Natab1[i-1]+(c1*K1a+c2*K2a+c3*K3a+c4*K4a)*dt
        Nbtab1[i]=Nbtab1[i-1]+(c1*K1b+c2*K2b+c3*K3b+c4*K4b)*dt
        Nctab1[i]=Nctab1[i-1]+(c1*K1c+c2*K2c+c3*K3c+c4*K4c)*dt
        
        i+1
    
    if Natab2[i-1]>0 or Nbtab2[i-1]>0:
        K1a=dNa2(Natab2[i-1])
        K2a=dNa2(Natab2[i-1]+a2*dt)
        K3a=dNa2(Natab2[i-1]+a3*dt)
        K4a=dNa2(Natab2[i-1]+a4*dt)
        K1b=dNb2(Natab2[i-1],       Nbtab2[i-1])
        K2b=dNb2(Natab2[i-1]+a2*dt, Nbtab2[i-1]+b21*K1b*dt)
        K3b=dNb2(Natab2[i-1]+a3*dt, Nbtab2[i-1]+b31*K1b*dt+b32*K2b*dt)
        K4b=dNb2(Natab2[i-1]+a4*dt, Nbtab2[i-1]+b41*K1b*dt+b42*K2b+b43*K3b*dt)   
        K1c=dNc2(Nbtab2[i-1])
        K2c=dNc2(Nbtab2[i-1]+a2*dt)
        K3c=dNc2(Nbtab2[i-1]+a3*dt)
        K4c=dNc2(Nbtab2[i-1]+a4*dt)

        Natab2[i]=Natab2[i-1]+(c1*K1a+c2*K2a+c3*K3a+c4*K4a)*dt
        Nbtab2[i]=Nbtab2[i-1]+(c1*K1b+c2*K2b+c3*K3b+c4*K4b)*dt
        Nctab2[i]=Nctab2[i-1]+(c1*K1c+c2*K2c+c3*K3c+c4*K4c)*dt
        
        i+1
        
    if Natab3[i-1]>0 or Nbtab3[i-1]>0:
        K1a=dNa3(Natab3[i-1])
        K2a=dNa3(Natab3[i-1]+a2*dt)
        K3a=dNa3(Natab3[i-1]+a3*dt)
        K4a=dNa3(Natab3[i-1]+a4*dt)
        K1b=dNb3(Natab3[i-1],       Nbtab3[i-1])
        K2b=dNb3(Natab3[i-1]+a2*dt, Nbtab3[i-1]+b21*K1b*dt)
        K3b=dNb3(Natab3[i-1]+a3*dt, Nbtab3[i-1]+b31*K1b*dt+b32*K2b*dt)
        K4b=dNb3(Natab3[i-1]+a4*dt, Nbtab3[i-1]+b41*K1b*dt+b42*K2b+b43*K3b*dt)
        K1c=dNc3(Nbtab3[i-1])
        K2c=dNc3(Nbtab3[i-1]+a2*dt)
        K3c=dNc3(Nbtab3[i-1]+a3*dt)
        K4c=dNc3(Nbtab3[i-1]+a4*dt)

        Natab3[i]=Natab3[i-1]+(c1*K1a+c2*K2a+c3*K3a+c4*K4a)*dt
        Nbtab3[i]=Nbtab3[i-1]+(c1*K1b+c2*K2b+c3*K3b+c4*K4b)*dt
        Nctab3[i]=Nctab3[i-1]+(c1*K1c+c2*K2c+c3*K3c+c4*K4c)*dt
        
        i+1
        
        
```

```python
plt.figure(figsize=(10,5))
plt.plot(ttab, Natab1, 'r',label="Na")
plt.plot(ttab, Nbtab1, 'b',label="Nb")
plt.plot(ttab, Nctab1, 'y',label="Nc")
plt.legend()

plt.figure(figsize=(10,5))
plt.plot(ttab, Natab2, 'r',label="Na")
plt.plot(ttab, Nbtab2, 'b',label="Nb")
plt.plot(ttab, Nctab2, 'y',label="Nc")
plt.legend()

plt.figure(figsize=(10,5))
plt.plot(ttab, Natab3, 'r',label="Na")
plt.plot(ttab, Nbtab3, 'b',label="Nb")
plt.plot(ttab, Nctab3, 'y',label="Nc")
plt.legend()
```
