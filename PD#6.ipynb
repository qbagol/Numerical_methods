```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = 2 
C1 = 1*10**-6 
C2 = 2*10**-6 
C3 = 3*10**-6 

RC1 = 1/(C1*R)
RC2 = 1/(C2*R)
RC3 = 1/(C3*R)

U0 = 3

dt = 0.0000001
tmax = 0.00001
t = np.arange(0,tmax,dt)

def f1(U):
    return RC1*(-U)
def f2(U):
    return RC2*(-U)
def f3(U):
    return RC3*(-U)

U1=np.zeros(len(t))
U2=np.zeros(len(t))
U3=np.zeros(len(t))
U1[0]=U0
U2[0]=U0
U3[0]=U0


c1,c2,c3,c4=1/6, 1/3, 1/3 ,1/6
a2,a3,a4=1/2,1/2,1
b21,b31,b32,b41,b42,b43=1/2,0,1/2,0,0,1

for i in range(len(t)):
    if U1[i-1]>0:
        K1=f1(U1[i-1])
        K2=f1(U1[i-1]+a2*dt)
        K3=f1(U1[i-1]+a3*dt)
        K4=f1(U1[i-1]+a4*dt)
        U1[i]=U1[i-1]+(c1*K1+c2*K2+c3*K3+c4*K4)*dt
        
for i in range(len(t)):
    if U2[i-1]>0:
        K1=f2(U2[i-1])
        K2=f2(U2[i-1]+a2*dt)
        K3=f2(U2[i-1]+a3*dt)
        K4=f2(U2[i-1]+a4*dt)
        U2[i]=U2[i-1]+(c1*K1+c2*K2+c3*K3+c4*K4)*dt

for i in range(len(t)):
    if U3[i-1]>0:
        K1=f3(U3[i-1])
        K2=f3(U3[i-1]+a2*dt)
        K3=f3(U3[i-1]+a3*dt)
        K4=f3(U3[i-1]+a4*dt)
        U3[i]=U3[i-1]+(c1*K1+c2*K2+c3*K3+c4*K4)*dt
        

plt.figure(figsize=(5,5))
plt.plot(U1,'r',label = "C= 1 mikroF R=2 omega") 
plt.plot(U2,'b',label = "C= 2 mikroF R=2 omega") 
plt.plot(U3,'y',label = "C= 3 mikroF R=2 omega") 
plt.legend()
```
